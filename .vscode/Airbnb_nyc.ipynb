Task 5 data analysis


*****. univeriate analysis

```python
#price distribution
data['price']
```

```python
# identifying outliers using boxplot
# limit data to prices below 1500 for better visualization
sea.boxplot(x=data['price'][data['price']<1500])
#show exact values of q1, q3, median, lower_bound, upper_bound
q1 = data['price'].quantile(0.25)
q3 = data['price'].quantile(0.75)
median = data['price'].median()
iqr = q3 - q1

print(f"Q1: {q1}, Q3: {q3}, Median: {median}, IQR: {iqr}")
```

```python
#histogram
plt.figure(figsize=(10,6))
sea.histplot(x=data['price'][data['price']<1500], bins=100, kde=True)
plt.xlim(0,1500)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
# anaylzing avilability_365 by histogram
plt.figure(figsize=(10,6))
sea.histplot(x=data['availability_365'][data['price']<1500], bins=30, kde=True)#there define where the data come from
plt.xlim(0,400)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Availability_365 Distribution')
plt.show()
```

```python
data.dtypes
```

```python
 # Analyze the neiborhood_group and price using panda groupby function
data.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

*** bi-variable analyze, -- how one column affect the other column

```python
data.columns
```

```python
#draw bar chart to visualize the average price in each neighbourhood_group,sub-category of neiborhood 
plt.figure(figsize=(10,6))
data.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Average Price in Each Neighbourhood Group')
#make the x labels more readable
plt.xticks(rotation=45)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()
```

```python
# how number_of_reviews affects price
plt.figure(figsize=(10,6))
sea.scatterplot(x='number_of_reviews', y='price', data=data[data['price']<1500], color='green')
plt.xlim(0,700)
plt.ylim(0,1500)
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.title('Number of Reviews vs Price')
plt.show()
#As the number of reviews increases, price tends to stay lower.
#In other words, cheaper listings are more popular and reviewed more often.
```

```python
#draw pairplot to visualize the relationship between multiple variables:price,minimum_nights,number_of_reviews,availability_365
sea.pairplot(data[['price','minimum_nights','number_of_reviews','availability_365']][data['price']<1500], diag_kind='kde', markers='+', plot_kws={'alpha':0.5, 'color':'purple'})
plt.suptitle('Pairplot of Price, Minimum Nights, Number of Reviews, Availability_365', y=1.02)
plt.show()
```

BY ABOVE:

price vs number_of_reviews

    More expensive listings tend to have fewer reviews.

    Cheaper listings are more likely to accumulate many reviews.

price vs minimum_nights

    Most listings have very low minimum stay requirements (1–30 days).

    A few listings require extremely high minimum stays (over 1000 days), which are outliers.

price vs availability_365

    Listings show various availability patterns, but 0 days and 365 days are the most common.

    There is no clear relationship between price and availability.

minimum_nights vs number_of_reviews

    Most listings require short stays, and their review counts cluster between 0–200.

    Listings with very long minimum stays usually have few reviews.

availability_365 vs number_of_reviews

    Listings available all year (365 days) show a wide distribution of reviews.

    Many listings available for 0 days or only part of the year have few reviews.

```python
# create a heatmap to visualize the correlation between multiple variables:price,
# number_of_reviews,minimum_nights,availability_365,revew__per_month.
plt.figure(figsize=(10,6))
#make x-labels more readable
plt.xticks(rotation=45)
correlation = data[['price','number_of_reviews','minimum_nights','availability_365','reviews_per_month']].corr()
sea.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

FROM ABOVE

The price of listings is mainly determined independently and shows no clear relationship with the number of reviews, minimum nights, or availability. However, the total number of reviews and reviews per month are closely related.
